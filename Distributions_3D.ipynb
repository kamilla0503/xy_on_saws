```python

fig = plt.figure(figsize=(13, 10))
ax = fig.add_subplot(111)


data_np = np.loadtxt("CPP_3D/Results/Iteration00/Counts_E_0.900000_0.000000_100.txt", skiprows=2)
plt.plot(data_np[:, 0]*0.001-2, data_np[:,1]/np.sum( data_np[:,1]) , "o", label = r"J = 0.9")

 
 
plt.title(r"Distribution of \langle e \rangle for N = 100 ", fontsize=22)

plt.xlabel(r'$ e$', fontsize = 22)
#plt.ylabel(r"$ U_4 $ ", fontsize=22)
plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 
#plt.ylim(-0.02, 0.7)

plt.grid()

plt.legend(loc="best", fontsize=20)


#plt.savefig("Thesis/Images/distr_energy_2500.png", bbox_inches='tight')
```
