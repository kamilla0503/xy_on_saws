```python
import matplotlib.pyplot as plt
import random
import numpy as np

import pandas as pd
from matplotlib.ticker import LogFormatter 
from scipy.optimize import curve_fit
from pathlib import Path
import matplotlib.colors
from matplotlib import ticker, cm
import pickle

import os                                                                   
import glob

import seaborn as sns 

from scipy.stats import norm
#%matplotlib notebook
```

```python
fig = plt.figure(figsize=(13, 10))
ax = fig.add_subplot(111)


files = sorted(list(glob.glob('CPP_3D/Results/Long/Counts_E*_700.txt')))
n = len(files)+1
#n=10

Js = [0.83 ,  0.85 ,  
       0.88 ,       
       0.898, 0.9  , 0.901, 0.904, 0.91 , 0.92   ]

#plt.set_cmap('gist_ncar')
custom_palette = sns.color_palette("Paired", n)

markers = ["d", "v", "s", "*", "^", "p", "o", "D", "*", "^", "d", "v"]

i=0
for file in files:   
    f=open(file)
    lines=f.readlines()
    J=float(lines[1].split()[1])
    if (J not in Js): continue
    color = plt.cm.hsv(i/(n-1))
    data_np = np.loadtxt(file, skiprows=2)
    plt.plot(data_np[:, 0]*0.001-2, data_np[:,1]/np.sum( data_np[:,1]) , markers[i], color=custom_palette[i], label = r"J = "+ str(J), ms = 8)
    i+=1
plt.legend(loc="best", fontsize=20)

plt.title(r"Distribution of  $ \langle e \rangle$ for N =  700 ", fontsize=22)

plt.xlabel(r'$  \langle e \rangle $', fontsize = 22)

plt.grid()

plt.legend(loc="best", fontsize=20)

plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 

plt.savefig("Thesis/Images/distr_energy_700.png", bbox_inches='tight') 
plt.savefig("/home/kamilla/SAW_paper_KF_1/img/distr_energy_700.png", bbox_inches='tight')
```
