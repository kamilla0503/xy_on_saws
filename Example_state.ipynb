```python
import matplotlib.pyplot as plt
import random
import numpy as np

import pandas as pd
from matplotlib.ticker import LogFormatter 
from scipy.optimize import curve_fit
from pathlib import Path
import matplotlib.colors
from matplotlib import ticker, cm
import pickle

from scipy.stats import norm
#%matplotlib notebook
```

```python
steps = [[1,0], [-1,0], [0,1], [0,-1]]

plt.figure(figsize=(14,14))

x = []
y = []

x.append(0)
y.append(0)

state = np.loadtxt("example_200.txt")
state[:,0] = np.array(state[:,0], dtype=int)

for i in range(len(state)-1):
    x.append(x[-1]+steps[int(state[i,0])][0])
    y.append(y[-1]+steps[int(state[i,0])][1])
    
    
plt.plot(x, y, '-', color = 'k', lw=3, alpha=0.3, zorder=1)
plt.scatter(x, y, c=state[:,1], s=205, edgecolors='k',  cmap='hsv', alpha = 1,zorder=2)    
    
plt.colorbar( )
plt.gca().set_aspect("equal")
plt.axis('off')
plt.savefig("Thesis/Images/state_example1.png")
```
