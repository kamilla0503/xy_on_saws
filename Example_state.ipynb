```python

plt.figure(figsize=(10,10))


ax = plt.axes()

ntheta = np.random.uniform(np.pi/2)
#print(np.cos(ntheta), np.sin(ntheta))
print( np.cos(ntheta)**2+np.sin(ntheta)**2  )
ax.arrow(0, 0, np.cos(ntheta)-0.045, np.sin(ntheta)-0.045, fc='red', ec='red', head_width=0.03,overhang = -0.2 )
ax.text(np.cos(ntheta)+0.05, np.sin(ntheta)+0.05, r"$\~n$", transform=ax.transData, size = 25)

spin = np.random.uniform(ntheta+np.pi/8, np.pi)
ax.arrow(0, 0, np.cos(spin), np.sin(spin), fc='lightblue', ec='black', head_width=0.03,overhang = -0.2 )
ax.text(np.cos(spin), np.sin(spin), r"$S_i$", transform=ax.transData, size = 25)
print(np.cos(spin)**2+np.sin(spin)**2)


c = -1.01
s = -(np.cos(ntheta)*c )/np.sin(ntheta)

 
dot_product = np.cos(spin)*np.cos(ntheta)+np.sin(spin)*np.sin(ntheta)
a = np.cos(spin)-2*dot_product*np.cos(ntheta)
b = np.sin(spin)-2*dot_product*np.sin(ntheta)
ax.arrow(0, 0, a, b, fc='lightblue', ec='black', head_width=0.03, overhang = -0.2)
print(a,b)
print(a**2+b**2)
ax.text(a+0.05, b+0.05, r"$S_i^{flipped}$", transform=ax.transData, size = 25)


plt.plot([0, c], [0, s], "--", alpha=0.8)

plt.grid()


plt.xlim(-1.15,0.75)
plt.ylim(-0.75,1.15)

plt.tick_params(axis='x', labelsize=22) 
plt.tick_params(axis='y', labelsize=22) 

plt.gca().set_aspect("equal")
#plt.axis('off')
#plt.savefig("Thesis/Images/cluster_flip.png")
```
